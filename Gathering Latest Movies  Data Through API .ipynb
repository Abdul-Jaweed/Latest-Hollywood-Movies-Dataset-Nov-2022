## Gathering Latest Movies Data Form Website "https://www.themoviedb.org/" Through API

```python
# import requests library and its work hitting url getting json data in python

import requests
import pandas as pd
```

```python
# fetching data

requests.get('https://api.themoviedb.org/3/movie/popular?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page=10')
```

```python
# fetching data into html format

requests.get('https://api.themoviedb.org/3/movie/popular?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page=1').text
```

```python
# storing json data in response variable

response = requests.get('https://api.themoviedb.org/3/movie/popular?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page=1').json()
```

```python
# fetching data using for loop

for i in response['results']:
    print(i['title'], i['vote_average'])
```

```python
# fetching all data from page 1 and convert into dataframe   

popularity = []
vote_count = []
original_language = []
title = []
vote_average = []
overview = []
release_date = []

for i in response['results']:
    
    popularity.append(i['popularity'])
    vote_count.append(i['vote_count'])
    original_language.append(i['original_language'])
    title.append(i['title'])
    vote_average.append(i['vote_average'])
    overview.append(i['overview'])
    release_date.append(i['release_date'])
    
    d = {'title':title, 'overview':overview,'original_language': original_language,
         'release_date':release_date, 'popularity':popularity, 'vote_count': vote_count,
         'vote_average':vote_average }
    
df = pd.DataFrame(d)
df
```

```python
# fetching all data from page 1 to 500 and convert into dataframe   

final = pd.DataFrame()

for j in range(1, 501):
    
    requests.get('https://api.themoviedb.org/3/movie/popular?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={}'.format(j)).json()
    
    popularity = []
    vote_count = []
    original_language = []
    title = []
    vote_average = []
    overview = []
    release_date = []

    for i in response['results']:

        popularity.append(i['popularity'])
        vote_count.append(i['vote_count'])
        original_language.append(i['original_language'])
        title.append(i['title'])
        vote_average.append(i['vote_average'])
        overview.append(i['overview'])
        release_date.append(i['release_date'])

    d = {'title':title, 'overview':overview,'original_language': original_language,
         'release_date':release_date, 'popularity':popularity, 'vote_count': vote_count,
         'vote_average':vote_average }

    df = pd.DataFrame(d)
    
    final = final.append(df, ignore_index=True)
```

```python
# printing data 

final
```

```python
# saving data to csv format

final.to_csv('latest_movies.csv')
```
